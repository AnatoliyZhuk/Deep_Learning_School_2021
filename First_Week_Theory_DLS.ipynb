<h1>Введение в машинное обучение <h1>

Машинное обучение делится на:
+ Unsupervised Learning (обучение без учителя) : Clustering (Кластеризация)
+ Reinforcement Learning (обучение с подкреплением) : Алгоритм за каждое свое действие получает награду или наказание. Алгоритм стремится научиться получать максимальную награду.
+ Supervised Learning (обучение c учителя) : Regression (Регрессия), Classification (Классификация) **Обучение без учителя** - у нас есть набор объектов, для которых мы не знаем ответ - их вообще не существует, нам нужно делать что-то только с самими данными <br/>

<h2> Unsupervised Learning (Обучение с учителем) <h2>

**Обучение с учителем** - один из способов машинного обучения, в ходе которого, используя набор объектов, для которых мы знаем ответ, мы пытаемся построить зависимость, которая наиболее точно предсказывает ответы для оставшегося набора объектов, для которых мы ответа не знаем <br/>

$Data$ - полная информация об объектах, которая была получена для решения задачи обучения с учителем <br/>
Разделим полный датасет $Data$ на:<br/>

+ $X$ - множество объектов с соответствующими признаками, которые необходимы для предсказания <br>
+ $Y$ - множество ответов (определенный признак объектов, который мы будет предсказывать) <br>
+ $y : X \rightarrow Y$ - функция, которая сопоставляет признаки объектов для предсказания и ответы <br><br>

Функцию $y$ мы можем называть: 
+ истинная зависимость
+ закон природы
+ функция, которую мы хотим восстановить 

Полный набор информации об объектах для предсказания $X$ можно представить в виде матрицы, где строки - это объекты, у каждого объекта есть признаки, которые характеризуют наш объект (один столбец - это один признак) <br><br>





\begin{equation*}
X = \left(
\begin{array}{cccc}
x_{11} & x_{12} & \ldots & x_{1m}\\
x_{21} & x_{22} & \ldots & x_{2m}\\
\vdots & \vdots & \ddots & \vdots\\
x_{n1} & x_{n2} & \ldots & x_{nm}
\end{array}
\right)
\quad
y = \left(
\begin{array}{cccc}
y_{1}\\
y_{2}\\
\vdots\\
x_{n}
\end{array}
\right)
\end{equation*}


Полный датасет может быть разделен на обучающий (тренировочный) датасет и тестовый датасет по объектам <br>
+ $X \rightarrow \quad X_{train} + X_{test}$ <br>
+ $y \rightarrow \quad y_{train} + y_{test}$ <br>


**Реальный кейс из практики** <br/>
**Люди** - объекты <br/>
**Ответы** - кликнет человек на рекламу или нет <br/>
**Функция** будет сопоставлять каждому человеку кликнет он на рекламу, которую мы ему покажем или нет <br/>
**Необходимость:** мы хотим показывать рекламу людям, которые хотят на нее кликнуть, потому что если показывать рекламу людям, которые не будут на нее кликать - это пустая трата денег <br/>
**Чем эффективнее мы будет предсказывать эту функцию, тем более эффективнее мы будем расходовать деньги** <br/>

**Типы признаков (features)**
+ Числовые (Numerical)
+ Категориальные (Categorical)
+ Ранжирование (Ordinal)


**Типы задач**
+ Regression (Регрессия) : множество объектов - это действительное число
+ Classification (Классификация) множество объектов - это класс
+ Ranking (ранжирование)

**Наша задача:** найти функцию хорошо приближающую реальную зависимость $y(x)$ <br/>
Назовем такое решение $\hat{y} : X \rightarrow Y$ - - эта функция должна быть вычислима на компьютере <br/><br/>

Заранее ограничим множество функций, из которых будем выбирать <br/>
$\digamma = (\hat{y_{\theta}} \, | \, \theta \in \Theta) \quad \Theta$ - множество параметров $\quad \digamma$ - параметризованное семейство<br/>

**Пример семейства моделей - функции порога** <br/>
**Задача:** Определить, можно ли ребенку пройти на атракцион, зная его рост и возраст <br/>
Множество, в котором мы будем искать решения состоит из функции вида:
$\hat{y}_{(a, b)}(x_1, x_2) = \begin{equation*}
 \begin{cases}
   1 \quad x_1 \geq a, x_2 \geq b
   \\
   0 \quad otherwise,
 \end{cases}
\end{equation*}$ <br/>

В данном случае **параметр** $\quad \theta = (a, b); \quad$  Множество возможных значений $\quad \Theta = R; \quad$ <br/>

Обучение - процесс выбора параметра $\theta$, которому соответствует наиболее подходящее нам решение задачи $\hat{y}_{\theta}(x_1, x_2)$

**Функция потерь (Loss function)** <br/>
$L(y,\hat{y}(x))$ - значение показывает насколько сильно наше предсказание отличается от реального значения <br/><br/>
Возможные функции потерь: <br/>
$L(y_{true},\hat{y}(x)) = (y_{true} - \hat{y}(x))^2$ - квадратичная функция потерь <br/>
$L(y_{true},\hat{y}(x)) = |y_{true} - \hat{y}(x)|$ - абсолютная функция потерь <br/>

**Эмперический риск** - cреднее значение функции потерь на обучающем датасете <br/>
Часто функцию эмперического риска называют **loss** <br/>
$\theta_{best} = argmin_{\theta \in \Theta} \frac{1}{dataset size} \sum\limits_{i} L(y_{true}^i,\hat{y}_{\theta}(x^i))$

<h1>Линейная регрессия<h1>

Стандартный вид линейной регрессии: <br/>
$\hat{y}(x_1, x_2, ..., x_n) = \theta_0 + \theta_1x_1 + \theta_2x_2 + ... + \theta_nx_n$ <br/>
Множество функций для линейной регрессии параметризуются весами <br/>
**Loss - Mean Squared Error:** $argmin_{\theta_0, ... \theta_n} \sum\limits_{i} (y_{true}^i - \theta_0 - \theta_1x_1^i - ... - \theta_nx_n^i)^2$

**Polynomial Regression** <br/>
Пусть у нас есть определенное количество объектов, которые описываются только одним признаком $x$ <br/>
Мы можем создать новые признаки по следующему правилу $x_1 = x, x_2 = x^2, ..., x_n = x^n$ <br/>
Тогда линейная регрессия от таких признаков называется **полиномиальной**:
$\hat{y}(x) = \theta_0 + \theta_1x + \theta_2x^2 + ... + \theta_nx^n$ <br/>

**Переобучение** - ситуация, когда качество на обучающей выборке сильно лучше качества, которое мы получим, применяя в реальности нашу модель <br/>
